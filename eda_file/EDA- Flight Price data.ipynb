```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("G:/Coding/Course related/EDA/flight_price.xlsx")
```

```python
#since route cannot have null value
df.dropna(axis=0, inplace=True) 
```

```python
df.isnull().sum()  
```

```python
df.columns
```

```python
df['Date_of_Journey']= pd.to_datetime(df['Date_of_Journey'])
```

```python
df['date'] = df['Date_of_Journey'].dt.day
df['month'] = df['Date_of_Journey'].dt.month
df['year'] = df['Date_of_Journey'].dt.year
```

```python
df.head()
```

```python
df['Arrival_Time']
```

```python
#extracting hour and minute from Arrival_Time
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df.Arrival_Time = df.Arrival_Time.astype(object)
```

```python
df['Arrival_Time']
```

```python
#now lets split the arrival time into hour and minute
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_mins'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
#converting hour and mins to int
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_mins'] = df['Arrival_mins'].astype(int)
```

```python
df.Dep_Time
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_mins'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_mins'] = df['Dep_mins'].astype(int)
```

```python
#we can drop the original columns
df.drop(['Date_of_Journey'],axis= 1, inplace = True)
df.drop(['Arrival_Time'], axis= 1, inplace = True)
df.drop(['Dep_Time'] , axis= 1, inplace = True)
```

```python
df['Total_Stops'].unique()
```

```python
#we can use label encoding for this since there is an order in the data
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df['Total_Stops']
```

```python
#we can drop the route column since we have the total steps and the destination and source
df.drop(['Route'], axis= 1, inplace = True) 
```

```python
df
```

```python
#now lets convert the duration into duration hours and duration minutes
df['Duration_hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_mins'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_mins'] = df['Duration_mins'].fillna(0)
df['Duration_mins'] = df['Duration_mins'].astype(int)
```

```python
df[['Airline', 'Source', 'Destination' , 'Additional_Info']]
```

```python
#since these all column are categorical we can use one hot encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded_data = encoder.fit_transform(df[['Airline', 'Source', 'Destination' , 'Additional_Info']]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
encoded_df = pd.DataFrame(encoded_data, columns =encoder.get_feature_names_out())
```

```python
df.drop(['Airline', 'Source', 'Destination' , 'Additional_Info'], axis= 1, inplace = True)
```

```python
final_df = pd.concat([df, encoded_df], axis = 1)
final_df
```

```python
final_df.drop(['Duration'], axis= 1, inplace = True) 
```
